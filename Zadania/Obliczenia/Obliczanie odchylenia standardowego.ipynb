# Obliczanie odchylenia standardowego (estymator obciążony)

## Definicja

Odchylenie standardowe mierzy, jak bardzo wartości w zbiorze danych różnią się od średniej. Estymator obciążony odchylenia standardowego, często nazywany po prostu odchyleniem standardowym populacji, jest obliczany przy założeniu, że dane obejmują całą populację, a nie tylko próbkę.

## Krok po kroku

1. **Zbiór danych**: 10, 11, 9

2. **Oblicz średnią (μ) zbioru danych**:
   - Średnia to suma wszystkich wartości podzielona przez liczbę wartości.
   - `μ = (10 + 11 + 9) / 3 = 30 / 3 = 10`

3. **Oblicz różnice między każdą wartością a średnią (xi - μ)**:
   - `10 - 10 = 0`
   - `11 - 10 = 1`
   - `9 - 10 = -1`

4. **Podnieś każdą różnicę do kwadratu (xi - μ)^2**:
   - `0^2 = 0`
   - `1^2 = 1`
   - `(-1)^2 = 1`

5. **Oblicz średnią arytmetyczną kwadratów różnic**:
   - `Średnia kwadratów różnic = (0 + 1 + 1) / 3 = 2 / 3 ≈ 0.6667`

6. **Oblicz odchylenie standardowe (σ)**:
   - Odchylenie standardowe to pierwiastek kwadratowy z tej średniej.
   - `σ = √0.6667 ≈ 0.8165`

## Podsumowanie

Odchylenie standardowe (estymator obciążony) dla zestawu danych `10, 11, 9` wynosi około `0.8165`.

### Przykład obliczeń w Pythonie

```python
import math

# Zbiór danych
data = [10, 11, 9]

# Obliczanie średniej
mean = sum(data) / len(data)

# Obliczanie sumy kwadratów różnic od średniej
squared_diff_sum = sum((x - mean) ** 2 for x in data)

# Obliczanie odchylenia standardowego
standard_deviation = math.sqrt(squared_diff_sum / len(data))
# Zaokrąglanie do dwoch miejsc po przecinku
standard_deviation_rounded = round(standard_deviation, 2)

print(f"Odchylenie standardowe (estymator obciążony) wynosi: {
      standard_deviation_rounded}")
```
